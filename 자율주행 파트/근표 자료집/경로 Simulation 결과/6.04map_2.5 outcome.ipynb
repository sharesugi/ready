```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# ====== 1. ê¸°ë³¸ ë°ì´í„° ë¡œë”© ======
maze = np.load("./maze.npy")
a_star_path = pd.read_csv("./a_star_path.csv")
actual_path = pd.read_csv("./tank_path0.csv")

# ====== 2. ì‹œìž‘ì ê³¼ ëª©ì ì§€ (Flask ì½”ë“œì—ì„œ ê°€ì ¸ì˜¨ ê°’) ======
start = (27, 50)
destination = (150, 275)

# ====== 3. ì¢Œí‘œ ì¶”ì¶œ ======
a_star_x = a_star_path["x"].tolist()
a_star_z = a_star_path["z"].tolist()

actual_x = actual_path["x"].tolist()
actual_z = actual_path["z"].tolist()

# ====== 4. ì‹œê°í™” ì‹œìž‘ ======
plt.figure(figsize=(10, 10))

# ðŸ”² ìž¥ì• ë¬¼ ì‹œê°í™” (mazeì—ì„œ 1ì¸ ì¢Œí‘œë§Œ ê·¸ë¦¼)
obstacle_coords = np.argwhere(maze == 1)
if len(obstacle_coords) > 0:
    obs_z, obs_x = zip(*obstacle_coords)
    plt.scatter(obs_x, obs_z, color='black', s=5, label="ìž¥ì• ë¬¼")

# ðŸ”µ A* ê²½ë¡œ
plt.plot(a_star_x, a_star_z, color='blue', linewidth=2, label="A* ê²½ë¡œ")

# ðŸ”¸ ì‹¤ì œ ì´ë™ ê²½ë¡œ
plt.plot(actual_x, actual_z, color='orange', linestyle='--', linewidth=2, label="ì‹¤ì œ ê²½ë¡œ")

# ðŸŸ¢ ì‹œìž‘ì 
plt.scatter([start[0]], [start[1]], c='green', s=100, marker='s', label="ì‹œìž‘ì ")

# ðŸ”´ ëª©ì ì§€
plt.scatter([destination[0]], [destination[1]], c='red', s=100, marker='X', label="ëª©ì ì§€")

# ====== 5. ê¾¸ë¯¸ê¸° ======
plt.title("A* ê²½ë¡œ, ì‹¤ì œ ê²½ë¡œ, ìž¥ì• ë¬¼")
plt.legend()
plt.grid(True)
plt.axis("equal")
plt.tight_layout()
plt.xlabel("X")
plt.ylabel("Z")
plt.imshow(maze, cmap=plt.cm.gray_r, origin='lower', alpha=0.2)  # ë°°ê²½ìœ¼ë¡œ maze íšŒìƒ‰ í‘œì‹œ
plt.show()

# ðŸ”µ A* ê²½ë¡œ â†’ blue line
# ðŸ”¸ ì‹¤ì œ ê²½ë¡œ â†’ ì£¼í™© ì ì„ 
# âš« ìž¥ì• ë¬¼ â†’ black dots
# ðŸŸ¢ ì‹œìž‘ì  â†’ green square
# ðŸ”´ ëª©ì ì§€ â†’ red X
```
