```python
import pandas as pd
import matplotlib.pyplot as plt

IP_files = ['true_hit_ratio_map2_YOLO.csv', 'true_hit_ratio_map4_YOLO.csv', 'true_hit_ratio_map5_YOLO.csv']

hit_ratios = []
labels = []

for file in IP_files:
    df = pd.read_csv(file)
    ratio = df['is_hit'].mean()
    hit_ratios.append(ratio)
    labels.append(file.replace('true_hit_ratio_', '').replace('.csv', ''))

plt.figure(figsize=(12, 5))
plt.bar(labels, hit_ratios, color=['#AEC6CF', '#B2DFDB', '#FFDAB9'])
plt.ylabel('Hit Ratio (Proportion of 1s)')
plt.title('Hit Ratio for Each CSV File')
plt.ylim(0, 1)

# 막대 위에 퍼센트 텍스트 표시
for i, ratio in enumerate(hit_ratios):
    plt.text(i, ratio - 0.05, f"{ratio*100:.2f}%", ha='center')

plt.show()
```
